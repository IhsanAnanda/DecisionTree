```python
#Importing Library yang akan digunakan
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from IPython.display import Image
import matplotlib.pyplot as plt
import pydotplus
import pandas as pd
import numpy as np
```

```python
#Load Dataset Iris dari library sklearn
iris = datasets.load_iris()
features = iris['data']
target = iris['target']
```

```python
#Mrnyiapkan Model
decisiontree = DecisionTreeClassifier(criterion="entropy",
                                      random_state=0, max_depth=10, 
                                      min_samples_split=2, min_samples_leaf=1, 
                                      min_weight_fraction_leaf=0, max_leaf_nodes=None, 
                                      min_impurity_decrease=0)
print("Model Siap Digunakan!")
```

```python
#Training Model
model = decisiontree.fit(features, target)
print("Training Selesai!")
```

```python
#Testing Model dengan 1 data
data1 = [[5, 2, 3, 4]]
hasil = model.predict(data1)
probabilitas = model.predict_proba(data1)
print("Hasil Klasifikasi : ", hasil)
print("Nilai Confidence : ", probabilitas)
```

```python
#Membuat visualisasi decision tree
dot_data = tree.export_graphviz(decisiontree, out_file=None,
                               feature_names=iris['feature_names'],
                               class_names=iris['target_names'])
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
graph.write_png("IrisDecisionTree.png")
print("Grafik Decision Tree Telah Diexport!")
```
